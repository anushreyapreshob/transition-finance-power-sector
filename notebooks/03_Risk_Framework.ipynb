# Risk & De-Risking Framework for Power Transition Financing

This notebook identifies the key risks faced by lenders and investors in financing the power sector transition and outlines commonly used de-risking mechanisms that enable large-scale capital mobilisation.


## Why Risk is Central to Transition Finance

While the capital requirement for the power transition is significant, the primary constraint is not availability of capital but the risk profile of projects. Understanding how risks are mitigated is critical to unlocking bank and institutional financing.


## Key Risks in Power Transition Projects

### Policy & Regulatory risk
* Tariff uncertainty
* Policy reversals
* Delays in approvals
### Counterparty Risk
* Payment delays by DISCOMs
* Weak credit profiles
### Technology Risk
* Performance risk in newer technologies
* Storage and grid integration uncertainty
### Construction & Execution Risk
* Delays and cost overruns
* EPC performance issues
### Market Risk
* Merchant price volatility
* Curtailment risk

## De-Risking Mechanisms Enabling Bankability
### Power Purchase Agreements (PPAs)
* Long-term revenue certainty
### Payment Security Mechanisms
* Letters of Credit
* Escrow accounts
* State guarantees
### Role of DFIs and Blended Finance
* First-loss capital
* Guarantees
* Subordinated debt
### Green Bonds and Sustainability-Linked Instruments
* Longer tenors
* Broader investor base
### Policy Stability and Planning Signals
* Clear renewable targets
* Grid expansion plans

```python
import pandas as pd

risk_derisking_map = pd.DataFrame({
    "Risk Category": [
        "Policy & Regulatory",
        "Counterparty",
        "Technology",
        "Construction",
        "Market"
    ],
    "Primary Concern for Lenders": [
        "Tariff uncertainty and regulatory change",
        "Delayed or defaulted payments",
        "Asset underperformance",
        "Delays and cost overruns",
        "Revenue volatility"
    ],
    "Key De-Risking Mechanism": [
        "Stable regulation and long-term tariffs",
        "Payment security mechanisms",
        "DFI participation and warranties",
        "EPC contracts and guarantees",
        "Long-term PPAs"
    ]
})

risk_derisking_map
```

**Note:** Risk levels are qualitative and illustrative (Low = 1, Medium = 2, High = 3), intended to highlight relative changes after de-risking rather than precise measurement.


```python
import matplotlib.pyplot as plt
import numpy as np

# Risk categories
risks = [
    "Policy & Regulatory",
    "Counterparty",
    "Technology",
    "Construction",
    "Market"
]

# Qualitative risk levels (Low=1, Medium=2, High=3)
before_derisking = [3, 3, 2, 2, 2]
after_derisking = [2, 2, 1, 1, 2]

data = np.array([before_derisking, after_derisking])

plt.figure(figsize=(8, 4))
plt.imshow(data)

plt.xticks(range(len(risks)), risks, rotation=30, ha='right')
plt.yticks([0, 1], ["Before De-Risking", "After De-Risking"])

plt.colorbar(label="Risk Level (1 = Low, 3 = High)")
plt.title("Qualitative Risk Heatmap: Impact of De-Risking Mechanisms")

plt.tight_layout()
plt.show()
```

### Interpretation

The heatmap illustrates how de-risking mechanisms significantly reduce technology and construction risks, while policy and counterparty risks remain partially structural and continue to influence financing decisions.
